### ¡Nosotros seguimos creando!
Empecemos importando este paquete (y numpy, porque lo usaremos en este Laboratorio también):

```python
import matplotlib.pyplot as plt
import numpy as np
```

Ahora, empecemos con la creación de un plano cartesiano vacío:

```python
fig, ax = plt.subplots()
```

Podemos modificar el tamaño del gráfico utilizando «figsize=(ancho, alto)»:

```python
fig, ax = plt.subplots(figsize=(10, 4))
```

Para crear una gráfica de función, tenemos que indicarle a Python que grafique puntos, es decir, cada punto del eje horizontal (abcsisa) debe corresponder con uno del eje vertical (ordenada). Dado que, nuevamente mencionamos, Python trabaja de manera discreta, tenemos que informarle cómo está dividida la abscisa: en cuántas divisiones. Mientras más divisiones tenga, más lisa (*smooth*) será la gráfica.

Para este ejemplo, vamos a graficar la función seno, vamos a observar 200 puntos y el rango será $(0,10)$:

```python
x = np.linspace(0, 10, 200)
y = np.sin(x)

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y, 'b-', linewidth=2)

plt.show()
```

Ahora vamos a observar solo 5 puntos. Recuerden que eso se logra con el atributo «linspace», que divide el rango dispuesto en partes iguales. Asimismo, cambiaremos el color azul (b) por verde (g) y en vez de una línea sólida (-), graficaremos una punteada (:):

```python
x = np.linspace(0, 10, 5)       
y = np.sin(x)

fig, ax = plt.subplots(figsize=(5, 3)) 
ax.plot(x, y, 'g:', linewidth=2)

plt.show()
```

#### Entendiendo matplotlib
Aquí estamos graficando un par de objetos:
- fig es una figura, como un lienzo vacío.
- ax es lo que «dibuja» los ejes donde se va a trazar las figuras, líneas, barras, etc.

Ahora veremos cómo agregar una **leyenda**, además, cambiaremos el color verde (g) por rojo (r), la haremos transparente hasta cierto grado (alpha) y cambiaremos el estilo de línea punteada (:) por discontinua (--):

```python
x = np.linspace(0, 10, 200)
y = np.sin(x)

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y, 'r--', linewidth=2, label='sine function', alpha=0.4)
ax.legend()

plt.show()
```

La ubicación de la leyenda puede modificarse utilizando «loc» dentro del atributo «legend». Note que si usa la opción «best», Python la colocará en una esquina donde no interfiera (i.e., se superponga) con los gráficos.

```python
x = np.linspace(0, 10, 200)
y = np.sin(x)

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y, 'r--', linewidth=2, label='sine function', alpha=0.4)
ax.legend(loc='upper center') # best, (upper, lower) x (left, center, right)

plt.show()
```

Ahora cambiaremos el color azul (b) por negro (k), cambiaremos el estilo de discontinua (--) por una de punto discontinuo (-.), agregaremos etiquetas y título, quitaremos la transparencia y modificaremos el rango de las ordenadas:

```python
x = np.linspace(0, 10, 200)
y = np.sin(x)

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y, 'k-.', linewidth=2, label='$y=\sin(x)$')
ax.legend(loc='upper center')
ax.set_yticks([-2, -1, 0, 1, 2])
ax.set_title('Ejemplo de gráfico')

plt.show()
```

A continuación, cambiaremos el color negro (k) por cian (c), agregaremos nombre a los ejes y agregaremos **tres** títulos diferentes en distintas posiciones. Finalmente, **guardaremos** el gráfico en PDF, en formato JPG y PNG en una carpeta *existente* llamada «output», para esto se reemplaza la última línea:

```python
x = np.linspace(0, 10, 200)
y = np.sin(x)

fig, ax = plt.subplots(figsize=(6, 4))
ax.plot(x, y, 'c-.', linewidth=2, label='$y=\sin(x)$')

ax.legend(loc='upper center')

ax.set_yticks([-2, -1, 0, 1, 2])

ax.set_title('Ejemplo de gráfico', fontsize=20)
plt.title('Título izquierdo', loc='left', fontsize=8)
plt.title('Título derecho', loc='right', fontsize=8)

ax.set_xlabel('abscisas' , fontsize=15)
ax.set_ylabel('ordenadas', fontsize=15)

plt.savefig("output/ejemplografico.pdf")
plt.savefig("output/ejemplografico.jpg")
plt.savefig("output/ejemplografico.png")
```

### Anotaciones y flechas indicativas
Aquí mostraremos una ejemplo de cómo incluir indicaciones y señales en gráficos.

```python
import pandas as pd
from datetime import datetime

date  = datetime(2008, 9, 15)
fig   = plt.figure(figsize=(16, 8))
ax    = fig.add_subplot(1, 1, 1)
dow   = pd.read_csv("data/dji.csv", index_col=0, parse_dates=True)
close = dow["Close"]
close.plot(ax=ax)
ax.annotate("Quiebra de Lehman",
            fontsize=20,
            xy=(date, close.loc[date] + 400), # hacia qué punto se señala (cabeza de la flecha)
            xytext=(date, 22000),             # la ubicación de la parte inicial inferior del texto
            arrowprops=dict(facecolor="red",
                            shrink=0.03))
ax.set_title("Dow Jones Industrial Average", size=30)
```

### Tipos de gráficos comunes
- ax.scatter(x, y): Crea un gráfico de dispersión de  x vs. y.
- ax.hist(x, bins): Crea un histograma.
- ax.fill_between(x, y, a): Crea un gráfico de x vs. y y rellena el gráfico entre a e y.

También podemos crear un gráfico que contenga 3 gráficas distintas, esto se logra con «plt.subplots(filas,columnas, figsize=(ancho, largo))». Por ejemplo, si filas=4 y columnas=3, tendremos un gráfico con 4 x 3 = 12 subgráficos. Para graficar cada uno tenemos que indicarle la ubicación respectiva, esto conserva la indexación que empieza de cero. Veremos un ejemplo en breves.

El atributo «subplots_adjust(alto,ancho)» funciona como control de espacio entre los sub-gráficos. Aquí mostramos algunos ejemplos:

```python
fig, ax = plt.subplots(1,3, figsize=(10, 4)) 

ax[0].hist([1, 2, 3, 4, 5, 4, 3, 2, 3, 4, 2, 3, 4, 4],bins=5, color="steelblue")
x = np.arange(0, 10, 0.1)
y = np.sin(x)
ax[1].fill_between(x, y, 0, color="seagreen")
ax[2].scatter(x, y)
plt.subplots_adjust(wspace=1)
#plt.subplots_adjust(wspace=0.7)
#plt.subplots_adjust(wspace=0.3)
#plt.subplots_adjust(wspace=0)
```

### Múltiples gráficos en un solo eje
A continuación, presentaremos tres densidades con distribuciones normales:

```python
from scipy.stats import norm
from random import uniform
```

```python
fig, ax = plt.subplots(figsize=(7, 3))
x       = np.linspace(-4, 4, 150)
for i in range(3):
    m, s = uniform(-1, 1), uniform(1, 2)
    y    = norm.pdf(x, loc=m, scale=s)
    current_label = f'$\mu = {m:.2}$'
    ax.plot(x, y, linewidth=2, alpha=0.6, label=current_label)
ax.legend()
plt.show()
```

### Múltiples sub-gráficos en una sola figura
A continuación, presentaremos seis histogramas. Para esto, tenemos que definir cómo los queremos distribuidos, en este caso, 3 filas x 2 columnas. 

```python
num_rows, num_cols = 3, 2
fig, axes = plt.subplots(num_rows, num_cols, figsize=(13, 12))
for i in range(num_rows):
    for j in range(num_cols):
        m, s = uniform(-1, 1), uniform(1, 2)
        x    = norm.rvs(loc=m, scale=s, size=100)
        axes[i, j].hist(x, alpha=0.6, bins=20)
        t    = f'$\mu = {m:.2}, \quad \sigma = {s:.2}$'
        axes[i, j].set(title=t, xticks=[-4, 0, 4], yticks=[])
plt.show()
```

### Gráficos en 3D
El atributo «Axes3D» del paquete mpl_toolkits nos permite crear gráfico tridimensionales. Además, es necesario el paquete «cm» de matplotlib.

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D
from matplotlib import cm
```

A continuación, proveemos un ejemplo:
$$\frac{\cos(x^2+y^2)}{1+x^2+y^2}$$

```python
def f(x, y):
    return np.cos(x**2 + y**2) / (1 + x**2 + y**2)

xgrid = np.linspace(-3, 3, 50)
ygrid = xgrid
x, y  = np.meshgrid(xgrid, ygrid)
fig   = plt.figure(figsize=(10, 6))
ax    = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, f(x, y), rstride=2, cstride=2, cmap=cm.jet, alpha=0.7, linewidth=0.25)
ax.set_zlim(-0.5, 1.0)
plt.show()
```
